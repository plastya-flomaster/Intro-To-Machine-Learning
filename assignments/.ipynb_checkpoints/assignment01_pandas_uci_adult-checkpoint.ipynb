```python
import numpy as np
import pandas as pd
pd.set_option('display.max.columns', 100)
# to draw pictures in jupyter notebook
%matplotlib inline 
import matplotlib.pyplot as plt
import seaborn as sns
# we don't like warnings
# you can comment the following 2 lines if you'd like to
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('adult_data.csv')
data.info()
```

**1. How many men and women (*sex* feature) are represented in this dataset?** 

```python
data['sex'].value_counts(normalize=True) #деляем в проценты
```

**2. What is the average age (*age* feature) of women?**

```python
age = data[data['sex'] == 'Female']['age'].mean()
print(f'Average women age is {round(age, 2)} years')
```

**3. What is the percentage of German citizens (*native-country* feature)?**

```python
percentage = data[data['native-country'] == 'Germany']['native-country'].count()/data.shape[0] * 10000
print(f'Here are {round(percentage,2)}% of German citizens')
```

**4-5. What are the mean and standard deviation of age for those who earn more than 50K per year (*salary* feature) and those who earn less than 50K per year? **

```python
rich = data[data['salary'].astype('str') == '>50K'];
poor = data[data['salary'].astype('str') == '<=50K']
print(f'Mean age of those who earn more than 50K: {round(rich.age.mean(), 2)} years \nMean age of those who earn less than 50K: {round(poor.age.mean(), 2)} years')
print('------------------------------------------------------------')
print(f'Std age of those who earn more than 50K: {round(rich.age.std(), 2)}\nMean age of those who earn less than 50K: {round(poor.age.std(), 2)}')

```

```python
#или
data.groupby('salary').age.agg([np.mean, np.std])
```

**6. Is it true that people who earn more than 50K have at least high school education? (*education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* or *Doctorate* feature)**

```python
edu = ['Bachelors', 'Prof-school', 'Assoc-acdm', 'Assoc-voc', 'Masters', 'Doctorate']

data[(data['salary'] == '<=50K') & (~data['education'].isin(edu))].tail()
```

**7. Display age statistics for each race (*race* feature) and each gender (*sex* feature). Use *groupby()* and *describe()*. Find the maximum age of men of *Amer-Indian-Eskimo* race.**

**8. Among whom is the proportion of those who earn a lot (>50K) greater: married or single men (*marital-status* feature)? Consider as married those who have a *marital-status* starting with *Married* (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.**

```python
#выберем богатых и женатых
rich = data[data['salary'] == '>50K']
married_count = rich[rich['marital-status'].apply(lambda x: x[:7] == 'Married')].shape[0]
single_count = rich.shape[0] - married_count

rich_men = rich[rich['sex'] == 'Male']
married_men_percent = rich_men[(rich_men['marital-status'].apply(lambda x: x[:7] == 'Married'))].shape[0]/married_count*100

print(f'rich people: {rich.shape[0]}')
print('------------')
print(f'rich married people: {married_count}\nrich single people: {single_count}')

print(f'rich men: {rich_men.shape[0]}')
print('------------')
print(f'rich married men: {round(married_men_percent, 2)}%\nrich single men: {round(100-married_men_percent, 2)}%')
single_men_percent
```

**9. What is the maximum number of hours a person works per week (*hours-per-week* feature)? How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?**

```python
a = data['hours-per-week'].max()
print(f'The maximum number of hours a person works per week is {a} hours')
```

```python
q = data[data['hours-per-week'] == a]
print(f'The number of people work {a} hours per week is {q.shape[0]} people')
```

```python
p = q[q['salary'] == '>50K'].shape[0]/q.shape[0]*100

print(f'And here {round(p, 2)}% earns a lot')
```
